```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Load data
df = pd.read_csv('../data/telco_comm_churn.csv')
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.describe()
```

## Handling missing values

```python
df.isnull().sum()
```

```python
# df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

df.fillna({'TotalCharges':df['TotalCharges'].median() }, inplace=True)
```

```python
df.isnull().sum()
```

Feature Engineering

#### Creating New Features

```python
df['tenure'].min(),  df['tenure'].max()
df['tenure']
```

```python
df['tenure_group'] = pd.cut(df['tenure'],
                            bins=[-1,12,24,48,73],
                            labels= [0,1,2,3],
                            include_lowest=True)
```

```python
df[['tenure', 'tenure_group']]
```

```python
df['tenure_group'].info()
```

```python
df['tenure_group'] = df['tenure_group'].astype(float).astype(int)
df['tenure_group'].info()
```

```python
df['avg_monthly_per_tenure'] = df['TotalCharges']/ (df['tenure'] + 1)
```
